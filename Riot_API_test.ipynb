```python
def filter(mts_dic, lane = 'NULL', champion = 0):
    matches = []
    for i in mts_dic['matches']:
        if ((lane == 'NULL') & (champion == 0)):
            matches.append(i['gameId'])
        elif (lane == 'NULL'):
            if (i['champion'] == champion):
                matches.append(i['gameId'])
        elif (champion == 0):
            if (i['lane'] == lane):
                matches.append(i['gameId'])            
        else :  
            if ((i['lane'] == lane) & (i['champion'] == champion)):
                matches.append(i['gameId'])
    return matches
```

```python
def remove_home_pts(x1,y1,x2,y2):
    print("BEFORE : blue_side = %d pts, red_side = %d pts " % (len(x1),len(x2)))
    for i in range(len(x1)-1,-1,-1):
        if (x1[i]<4000 and y1[i]<4000):
            del x1[i]
            del y1[i]
    for i in range(len(x2)-1,-1,-1):
        if (x2[i]>11000 and y2[i]>11000):
            del x2[i]
            del y2[i]
    print("AFTER  : blue_side = %d pts, red_side = %d pts " % (len(x1),len(x2)))
    return x1,y1,x2,y2
```

```python
def heatmap(x, y, s, bins):
    heatmap, xedge, yedge = np.histogram2d(x, y, range = [[0, 15000], [0, 15000]], bins=bins)    
    heatmap = gaussian_filter(heatmap, sigma=s)
    return heatmap.T

def show_heat_scatter(x,y,s, bins = 500):
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    fig = plt.figure(figsize=(5,10))
    extent = [0,15000,0,15000]
    
    axs[0].plot(x, y, 'b.', markersize=2)
    axs[0].set_xlim([0, 15000])
    axs[0].set_ylim([0, 15000])
    axs[0].set_title("Scatter plot")
    img = plt.imread("./Minimap.png")
    axs[0].imshow(img, zorder=0, extent=extent)

    heat = heatmap(x, y, s, bins)
    axs[1].imshow(heat, extent = extent, origin='lower', cmap=cm.jet)
    
    axs[1].set_title("Smoothing with  $\sigma$ = %d" % s)
    plt.show()
```
